```python
df['Category'] = df['Category'].astype('category')
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format="%d/%m/%Y")
```

```python
# 1) Order Date ve Ship Date sütunlarını datetime’a çevir
df['Order Date'] = pd.to_datetime(
    df['Order Date'],
    dayfirst=True,         # eğer dd/mm/yyyy formatındaysa
    errors='coerce'        # parse edilemeyenleri NaT yapar
)

df['Ship Date']  = pd.to_datetime(
    df['Ship Date'],
    dayfirst=True,
    errors='coerce'
)

# 2) Yıl/ay/gün bilgilerini yeni sütunlara al
df['Order Year']  = df['Order Date'].dt.year
df['Order Month'] = df['Order Date'].dt.month
df['Order Day']   = df['Order Date'].dt.day

df['Ship Year']   = df['Ship Date'].dt.year
df['Ship Month']  = df['Ship Date'].dt.month
df['Ship Day']    = df['Ship Date'].dt.day
```

```python
city_sales = df.groupby("City")["Sales"].sum().sort_values(ascending=False).head(10)
print(city_sales)

# Bar grafiği
city_sales.plot(kind="bar", figsize=(10,6), color="#3498db")
plt.title("Top 10 Cities with the Highest Sales")
plt.ylabel("Sales Amount ($)")
plt.xlabel("City")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
print(category_sales)

sub_category_sales = df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False)
print(sub_category_sales)

# Category bar chart
category_sales.plot(kind='bar', color='#e74c3c')
plt.title('Sales by Category')
plt.ylabel('Sales Amount (USD)')
plt.show()
```

```python
segment_sales = df.groupby('Segment')['Sales'].sum()
print(segment_sales)

# Pie chart
segment_sales.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(8,8))
plt.title('Sales Distribution by Customer Segment')
plt.ylabel('')
plt.show()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

monthly_sales = df.groupby(df['Order Date'].dt.to_period('M'))['Sales'].sum()

# Line plot
monthly_sales.plot(figsize=(12,6), marker='o')
plt.title('Monthly Sales Trend')
plt.ylabel('Sales Amount (USD)')
plt.xlabel('Month')
plt.grid()
plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)
print(region_sales)

# Bar chart
region_sales.plot(kind='bar', color='#2ecc71')
plt.title('Sales by Region')
plt.ylabel('Sales Amount (USD)')
plt.show()
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(5)
print(top_products)

# Bar chart
top_products.plot(kind='bar', color='#9b59b6')
plt.title('Top 5 Best Selling Products')
plt.ylabel('Sales Amount (USD)')
plt.show()
```
