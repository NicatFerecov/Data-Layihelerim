```python
# Clean up author names (remove extra spaces, fix formatting)
books['authors'] = books['authors'].str.strip()
books['authors'] = books['authors'].str.replace(',', ' &')

# Extract main title (remove series info in parentheses)
books['main_title'] = books['title'].str.replace(r'\s*\([^)]*\)', '', regex=True)

# Convert to title case
books['title'] = books['title'].str.title()
books['authors'] = books['authors'].str.title()
```

```python
# Count characters in titles
books['title_length'] = books['title'].str.len()

# Count words in titles
books['title_word_count'] = books['title'].str.split().str.len()

# Count characters in author names
books['author_length'] = books['authors'].str.len()
```

```python
# Extract series information
books['series'] = books['title'].str.extract(r'\(([^)]*#[^)]*)\)')

# Check if title contains series info
books['has_series'] = books['title'].str.contains(r'\(.*#.*\)', na=False)

# Extract first author only
books['first_author'] = books['authors'].str.split(',').str[0]
```

```python
# Find books with specific words in title
harry_potter_books = books[books['title'].str.contains('Harry Potter', na=False)]

# Find books by specific authors
collins_books = books[books['authors'].str.contains('Collins', na=False)]

# Books published in 21st century
modern_books = books[books['pub_year'] >= 2000]
```

```python
# Create author count
books['author_count'] = books['authors'].str.count(',') + 1

# Normalize titles (lowercase, remove punctuation)
books['title_normalized'] = books['title'].str.lower().str.replace(r'[^\w\s]', '', regex=True)

# Extract publication decade as string
books['decade_str'] = (books['pub_year'] // 10 * 10).astype(str) + 's'
```

```python
# Multiple conditions using vectorized operations
high_rated_popular = books[
    (books['avg_rating'] >= 4.0) &
    (books['total_ratings'] >= 1000000) &
    (books['pub_year'] >= 2000)
]

# Use .query() for complex conditions (often faster)
modern_classics = books.query(
    'avg_rating >= 4.2 and total_ratings >= 500000 and pub_year >= 1950'
)
```

```python
# Pre-compile regex patterns for repeated use
import re
fantasy_pattern = re.compile(r'harry potter|twilight|hunger games', re.IGNORECASE)
books['is_popular_series'] = books['title'].str.contains(fantasy_pattern, na=False)

# Use categorical data for faster filtering
books['rating_category'] = pd.Categorical(
    books['avg_rating'].apply(lambda x: 'Excellent' if x >= 4.5 else 'Good' if x >= 4.0 else 'Fair'),
    categories=['Fair', 'Good', 'Excellent'],
    ordered=True
)

# Filter using categorical (faster than string comparison)
excellent_books = books[books['rating_category'] == 'Excellent']
```

```python
# Use .loc for specific columns to reduce memory
essential_columns = ['title', 'authors', 'avg_rating', 'total_ratings']
filtered_books = books.loc[
    (books['avg_rating'] >= 4.0) & (books['total_ratings'] >= 100000),
    essential_columns
]

# Chain operations efficiently
top_books = (books
    .query('avg_rating >= 4.30 and total_ratings >= 100000')
    .nlargest(10, 'total_ratings')
    [['title', 'authors', 'avg_rating', 'total_ratings']]
)
```

```python
# Author evolution analysis
def author_evolution(author_data):
    if len(author_data) < 2:
        return pd.Series({'evolution': 'insufficient_data'})

    # Sort by publication year
    sorted_data = author_data.sort_values('pub_year')

    # Calculate trends
    rating_trend = np.polyfit(range(len(sorted_data)), sorted_data['avg_rating'], 1)[0]
    popularity_trend = np.polyfit(range(len(sorted_data)), sorted_data['total_ratings'], 1)[0]

    return pd.Series({
        'books_count': len(sorted_data),
        'career_span': sorted_data['pub_year'].max() - sorted_data['pub_year'].min(),
        'rating_trend': rating_trend,'popularity_trend': popularity_trend,
        'peak_book': sorted_data.loc[sorted_data['avg_rating'].idxmax(), 'title'],
        'latest_book': sorted_data.iloc[-1]['title']
    })

author_evolution_analysis = books.groupby('authors').apply(author_evolution)
```
