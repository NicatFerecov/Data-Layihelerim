```python
#  10. Custom Filtering Functions
print("\n10. CUSTOM FILTERING FUNCTIONS")
print("-" * 30)

def is_blockbuster(row):
    """Define blockbuster as high revenue AND high votes"""
    return (row['Revenue (Millions)'] >= 100) and (row['Votes'] >= 50000)

def is_critically_acclaimed(row):
    """Define critically acclaimed as high rating AND high metascore"""
    return (row['Rating'] >= 8.0) and (row['Metascore'] >= 80)

# Apply custom filters
blockbusters = df[df.apply(is_blockbuster, axis=1)]
critically_acclaimed = df[df.apply(is_critically_acclaimed, axis=1)]

print(f"Blockbuster movies: {len(blockbusters)} movies")
print(f"Critically acclaimed movies: {len(critically_acclaimed)} movies")

# 11. Filtering with Statistical Measures
print("\n11. STATISTICAL FILTERING")
print("-" * 30)

# Movies above average in multiple categories
above_avg_rating = df[df['Rating'] > df['Rating'].mean()]
above_avg_revenue = df[df['Revenue (Millions)'] > df['Revenue (Millions)'].mean()]
above_avg_votes = df[df['Votes'] > df['Votes'].mean()]

print(f"Above average rating: {len(above_avg_rating)} movies")
print(f"Above average revenue: {len(above_avg_revenue)} movies")
print(f"Above average votes: {len(above_avg_votes)} movies")

# Movies that are above average in ALL categories
above_avg_all = df[
    (df['Rating'] > df['Rating'].mean()) &
    (df['Revenue (Millions)'] > df['Revenue (Millions)'].mean()) &
    (df['Votes'] > df['Votes'].mean()) &
    (df['Metascore'] > df['Metascore'].mean())
]
print(f"Above average in ALL metrics: {len(above_avg_all)} movies")

# 12. Filtering and Immediate Analysis
print("\n12. FILTERING WITH IMMEDIATE ANALYSIS")
print("-" * 40)

# Filter and analyze high-rated movies
high_rated_analysis = df[df['Rating'] >= 8.0]
print(f"High-rated movies analysis:")
print(f"  Count: {len(high_rated_analysis)}")
print(f"  Average Revenue: ${high_rated_analysis['Revenue (Millions)'].mean():.1f}M")
print(f"  Average Runtime: {high_rated_analysis['Runtime (Minutes)'].mean():.0f} minutes")
print(f"  Most common genre: {high_rated_analysis['Genre'].mode().iloc[0]}")

print("\n" + "="*60)
print("FILTERING BEST PRACTICES")
print("="*60)

print("""
1. Use parentheses for complex conditions: (condition1) & (condition2)
2. Use & for AND, | for OR, ~ for NOT
3. For string columns, use .str methods: .contains(), .startswith(), .endswith()
4. Use .isin() for multiple value matching
5. Use .between() for range filtering
6. Use .query() for readable complex conditions
7. Consider using .loc[] for label-based filtering
8. Chain filters step by step for complex logic
9. Use quantiles for relative filtering
10. Always check data types before filtering
""")

# Save filtered results example
print("\nSaving filtered results:")
print("# Save high-rated movies to CSV")
high_rated_movies = df[df['Rating'] >= 8.0]
high_rated_movies.to_csv('high_rated_movies.csv', index=False)

print("\n# Save multiple filtered datasets")
print("filters = {")
print("    'action_movies': df[df['Genre'] == 'Action'],")
print("    'recent_movies': df[df['Year'] >= 2020],")
print("    'blockbusters': df[df['Revenue (Millions)'] >= 100]")
print("}")
print("for name, filtered_df in filters.items():")
print("    filtered_df.to_csv(f'{name}.csv', index=False)")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.offline as pyo
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Set style for better-looking plots
plt.style.use('dark_background')
sns.set_palette("husl")

class MovieDataVisualizer:
    def __init__(self, df=None):
        if df is None:
            self.df = self.create_sample_data()
        else:
            self.df = df.copy()

        # Set up color palettes
        self.colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#96CEB4', '#FECA57', '#FF9FF3', '#54A0FF']
        self.genre_colors = {
            'Action': '#FF6B6B',
            'Drama': '#4ECDC4',
            'Comedy': '#FECA57',
            'Thriller': '#FF9FF3',
            'Horror': '#54A0FF',
            'Sci-Fi': '#96CEB4',
            'Romance': '#DDA0DD'
        }

    def create_sample_data(self):
        """Create sample movie dataset for demonstration"""
        np.random.seed(42)
        genres = ['Action', 'Drama', 'Comedy', 'Thriller', 'Horror', 'Sci-Fi', 'Romance']

        data = {
            'Rank': range(1, 501),
            'Title': [f'Movie {i}' for i in range(1, 501)],
            'Genre': np.random.choice(genres, 500),
            'Director': [f'Director {i}' for i in range(1, 501)],
            'Year': np.random.randint(1990, 2025, 500),
            'Runtime (Minutes)': np.random.randint(80, 180, 500),
            'Rating': np.round(np.random.uniform(5.0, 9.5, 500), 1),
            'Votes': np.random.randint(1000, 100000, 500),
            'Revenue (Millions)': np.round(np.random.uniform(1, 500, 500), 1),
            'Metascore': np.random.randint(40, 95, 500)
        }

        return pd.DataFrame(data)

    def apply_filters(self, min_rating=5, max_rating=10, min_year=1990, max_year=2024,
                     genres=None, min_revenue=0, max_revenue=500, min_runtime=80,
                     max_runtime=180, min_votes=1000, max_votes=100000):
        """Apply filters to the dataset"""
        filtered_df = self.df.copy()

        # Apply filters
        filtered_df = filtered_df[
            (filtered_df['Rating'] >= min_rating) &
            (filtered_df['Rating'] <= max_rating) &
            (filtered_df['Year'] >= min_year) &
            (filtered_df['Year'] <= max_year) &
            (filtered_df['Revenue (Millions)'] >= min_revenue) &
            (filtered_df['Revenue (Millions)'] <= max_revenue) &
            (filtered_df['Runtime (Minutes)'] >= min_runtime) &
            (filtered_df['Runtime (Minutes)'] <= max_runtime) &
            (filtered_df['Votes'] >= min_votes) &
            (filtered_df['Votes'] <= max_votes)
        ]

        if genres:
            filtered_df = filtered_df[filtered_df['Genre'].isin(genres)]

        return filtered_df

    def create_matplotlib_dashboard(self, filtered_df=None):
        """Create comprehensive matplotlib dashboard"""
        if filtered_df is None:
            filtered_df = self.df

        # Create figure with subplots
        fig = plt.figure(figsize=(20, 16))
        fig.suptitle('Movie Data Analysis Dashboard', fontsize=24, fontweight='bold', color='white')

        # 1. Genre distribution (Bar chart)
        ax1 = plt.subplot(3, 3, 1)
        genre_counts = filtered_df['Genre'].value_counts()
        bars = ax1.bar(genre_counts.index, genre_counts.values, color=self.colors[:len(genre_counts)])
        ax1.set_title('Movies by Genre', fontsize=14, fontweight='bold')
        ax1.set_xlabel('Genre')
        ax1.set_ylabel('Count')
        plt.xticks(rotation=45)

        # Add value labels on bars
        for bar in bars:
            height = bar.get_height()
            ax1.text(bar.get_x() + bar.get_width()/2., height,
                    f'{int(height)}', ha='center', va='bottom')

        # 2. Rating distribution (Histogram)
        ax2 = plt.subplot(3, 3, 2)
        ax2.hist(filtered_df['Rating'], bins=20, color='skyblue', alpha=0.7, edgecolor='black')
        ax2.axvline(filtered_df['Rating'].mean(), color='red', linestyle='--',
                   label=f'Mean: {filtered_df["Rating"].mean():.1f}')
        ax2.set_title('Rating Distribution', fontsize=14, fontweight='bold')
        ax2.set_xlabel('Rating')
        ax2.set_ylabel('Frequency')
        ax2.legend()

        # 3. Revenue vs Rating (Scatter plot)
        ax3 = plt.subplot(3, 3, 3)
        genres_unique = filtered_df['Genre'].unique()
        for i, genre in enumerate(genres_unique):
            genre_data = filtered_df[filtered_df['Genre'] == genre]
            ax3.scatter(genre_data['Rating'], genre_data['Revenue (Millions)'],
                       alpha=0.6, label=genre, color=self.colors[i % len(self.colors)])
        ax3.set_title('Revenue vs Rating by Genre', fontsize=14, fontweight='bold')
        ax3.set_xlabel('Rating')
        ax3.set_ylabel('Revenue (Millions)')
        ax3.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

        # 4. Movies by Year (Line plot)
        ax4 = plt.subplot(3, 3, 4)
        year_counts = filtered_df['Year'].value_counts().sort_index()
        ax4.plot(year_counts.index, year_counts.values, marker='o', linewidth=2, markersize=4)
        ax4.set_title('Movies by Year', fontsize=14, fontweight='bold')
        ax4.set_xlabel('Year')
        ax4.set_ylabel('Number of Movies')
        ax4.grid(True, alpha=0.3)

        # 5. Runtime distribution by Genre (Box plot)
        ax5 = plt.subplot(3, 3, 5)
        genre_list = filtered_df['Genre'].unique()
        runtime_by_genre = [filtered_df[filtered_df['Genre'] == genre]['Runtime (Minutes)'].values
                           for genre in genre_list]
        bp = ax5.boxplot(runtime_by_genre, labels=genre_list, patch_artist=True)
        for patch, color in zip(bp['boxes'], self.colors):
            patch.set_facecolor(color)
        ax5.set_title('Runtime Distribution by Genre', fontsize=14, fontweight='bold')
        ax5.set_xlabel('Genre')
        ax5.set_ylabel('Runtime (Minutes)')
        plt.xticks(rotation=45)

        # 6. Revenue distribution (Histogram)
        ax6 = plt.subplot(3, 3, 6)
        ax6.hist(filtered_df['Revenue (Millions)'], bins=30, color='lightgreen', alpha=0.7, edgecolor='black')
        ax6.axvline(filtered_df['Revenue (Millions)'].mean(), color='red', linestyle='--',
                   label=f'Mean: ${filtered_df["Revenue (Millions)"].mean():.1f}M')
        ax6.set_title('Revenue Distribution', fontsize=14, fontweight='bold')
        ax6.set_xlabel('Revenue (Millions)')
        ax6.set_ylabel('Frequency')
        ax6.legend()

        # 7. Correlation heatmap
        ax7 = plt.subplot(3, 3, 7)
        numeric_cols = ['Rating', 'Runtime (Minutes)', 'Revenue (Millions)', 'Votes', 'Metascore', 'Year']
        correlation_matrix = filtered_df[numeric_cols].corr()
        im = ax7.imshow(correlation_matrix, cmap='coolwarm', aspect='auto', vmin=-1, vmax=1)
        ax7.set_xticks(range(len(numeric_cols)))
        ax7.set_yticks(range(len(numeric_cols)))
        ax7.set_xticklabels(numeric_cols, rotation=45)
        ax7.set_yticklabels(numeric_cols)
        ax7.set_title('Correlation Matrix', fontsize=14, fontweight='bold')

        # Add correlation values to heatmap
        for i in range(len(numeric_cols)):
            for j in range(len(numeric_cols)):
                text = ax7.text(j, i, f'{correlation_matrix.iloc[i, j]:.2f}',
                               ha="center", va="center", color="black", fontweight='bold')

        # 8. Top 10 highest-rated movies
        ax8 = plt.subplot(3, 3, 8)
        top_movies = filtered_df.nlargest(10, 'Rating')[['Title', 'Rating']]
        bars = ax8.barh(range(len(top_movies)), top_movies['Rating'], color='gold')
        ax8.set_yticks(range(len(top_movies)))
        ax8.set_yticklabels([f"{title[:15]}..." if len(title) > 15 else title
                            for title in top_movies['Title']], fontsize=8)
        ax8.set_xlabel('Rating')
        ax8.set_title('Top 10 Highest Rated Movies', fontsize=14, fontweight='bold')

        # Add value labels
        for i, bar in enumerate(bars):
            width = bar.get_width()
            ax8.text(width, bar.get_y() + bar.get_height()/2,
                    f'{width:.1f}', ha='left', va='center')

        # 9. Summary statistics
        ax9 = plt.subplot(3, 3, 9)
        ax9.axis('off')
        stats_text = f"""
        SUMMARY STATISTICS

        Total Movies: {len(filtered_df):,}

        Average Rating: {filtered_df['Rating'].mean():.2f}
        Average Revenue: ${filtered_df['Revenue (Millions)'].mean():.1f}M
        Average Runtime: {filtered_df['Runtime (Minutes)'].mean():.0f} min

        Top Genre: {filtered_df['Genre'].value_counts().index[0]}
        Most Recent Year: {filtered_df['Year'].max()}
        Oldest Year: {filtered_df['Year'].min()}

        Highest Revenue: ${filtered_df['Revenue (Millions)'].max():.1f}M
        Highest Rating: {filtered_df['Rating'].max():.1f}
        """
        ax9.text(0.1, 0.9, stats_text, transform=ax9.transAxes, fontsize=12,
                verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

        plt.tight_layout()
        return fig

    def create_seaborn_advanced_plots(self, filtered_df=None):
        """Create advanced seaborn visualizations"""
        if filtered_df is None:
            filtered_df = self.df

        fig, axes = plt.subplots(2, 2, figsize=(16, 12))
        fig.suptitle('Advanced Statistical Visualizations', fontsize=20, fontweight='bold')

        # 1. Pairplot-style correlation
        ax1 = axes[0, 0]
        sns.scatterplot(data=filtered_df, x='Rating', y='Revenue (Millions)',
                       hue='Genre', alpha=0.7, ax=ax1)
        ax1.set_title('Rating vs Revenue by Genre')

        # 2. Violin plot for ratings by genre
        ax2 = axes[0, 1]
        sns.violinplot(data=filtered_df, x='Genre', y='Rating', ax=ax2)
        ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45)
        ax2.set_title('Rating Distribution by Genre')

        # 3. Heatmap of average metrics by genre and decade
        ax3 = axes[1, 0]
        filtered_df['Decade'] = (filtered_df['Year'] // 10) * 10
        pivot_data = filtered_df.groupby(['Genre', 'Decade'])['Rating'].mean().unstack(fill_value=0)
        sns.heatmap(pivot_data, annot=True, cmap='viridis', ax=ax3, fmt='.1f')
        ax3.set_title('Average Rating by Genre and Decade')

        # 4. Joint plot for runtime vs revenue
        ax4 = axes[1, 1]
        sns.scatterplot(data=filtered_df, x='Runtime (Minutes)', y='Revenue (Millions)',
                       hue='Genre', alpha=0.7, ax=ax4)
        ax4.set_title('Runtime vs Revenue by Genre')

        plt.tight_layout()
        return fig

    def create_plotly_interactive_dashboard(self, filtered_df=None):
        """Create interactive Plotly dashboard"""
        if filtered_df is None:
            filtered_df = self.df

        # Create subplots
        fig = make_subplots(
            rows=3, cols=2,
            subplot_titles=('Genre Distribution', 'Rating vs Revenue',
                          'Movies by Year', 'Runtime Distribution',
                          'Revenue Distribution', 'Top Movies by Rating'),
            specs=[[{"type": "bar"}, {"type": "scatter"}],
                   [{"type": "scatter"}, {"type": "histogram"}],
                   [{"type": "histogram"}, {"type": "bar"}]]
        )

        # 1. Genre distribution
        genre_counts = filtered_df['Genre'].value_counts()
        fig.add_trace(
            go.Bar(x=genre_counts.index, y=genre_counts.values,
                   marker_color=self.colors[:len(genre_counts)],
                   name='Genre Count'),
            row=1, col=1
        )

        # 2. Rating vs Revenue scatter
        for i, genre in enumerate(filtered_df['Genre'].unique()):
            genre_data = filtered_df[filtered_df['Genre'] == genre]
            fig.add_trace(
                go.Scatter(x=genre_data['Rating'], y=genre_data['Revenue (Millions)'],
                          mode='markers', name=genre,
                          marker=dict(color=self.colors[i % len(self.colors)], size=8, opacity=0.7),
                          text=genre_data['Title'],
                          hovertemplate='<b>%{text}</b><br>Rating: %{x}<br>Revenue: $%{y}M<extra></extra>'),
                row=1, col=2
            )

        # 3. Movies by year
        year_counts = filtered_df['Year'].value_counts().sort_index()
        fig.add_trace(
            go.Scatter(x=year_counts.index, y=year_counts.values,
                      mode='lines+markers', name='Movies per Year',
                      line=dict(color='lightblue', width=3)),
            row=2, col=1
        )

        # 4. Runtime distribution
        fig.add_trace(
            go.Histogram(x=filtered_df['Runtime (Minutes)'], nbinsx=20,
                        marker_color='lightgreen', name='Runtime Distribution'),
            row=2, col=2
        )

        # 5. Revenue distribution
        fig.add_trace(
            go.Histogram(x=filtered_df['Revenue (Millions)'], nbinsx=30,
                        marker_color='orange', name='Revenue Distribution'),
            row=3, col=1
        )

        # 6. Top movies by rating
        top_movies = filtered_df.nlargest(10, 'Rating')
        fig.add_trace(
            go.Bar(y=top_movies['Title'], x=top_movies['Rating'],
                   orientation='h', marker_color='gold', name='Top Rated'),
            row=3, col=2
        )

        # Update layout
        fig.update_layout(
            height=1200,
            title_text="Interactive Movie Data Dashboard",
            title_x=0.5,
            title_font_size=24,
            showlegend=False,
            template="plotly_dark"
        )

        return fig

    def create_custom_filter_visualization(self, **filter_kwargs):
        """Create visualization with custom filters"""
        filtered_df = self.apply_filters(**filter_kwargs)

        print(f"Filtered dataset contains {len(filtered_df)} movies")
        print(f"Filter applied: {filter_kwargs}")

        # Create the dashboard
        fig = self.create_matplotlib_dashboard(filtered_df)
        return fig, filtered_df

    def analyze_filter_impact(self, filter_sets):
        """Analyze the impact of different filter combinations"""
        results = []

        for name, filters in filter_sets.items():
            filtered_df = self.apply_filters(**filters)
            results.append({
                'Filter Name': name,
                'Movie Count': len(filtered_df),
                'Avg Rating': filtered_df['Rating'].mean() if len(filtered_df) > 0 else 0,
                'Avg Revenue': filtered_df['Revenue (Millions)'].mean() if len(filtered_df) > 0 else 0,
                'Top Genre': filtered_df['Genre'].mode().iloc[0] if len(filtered_df) > 0 else 'N/A'
            })

        comparison_df = pd.DataFrame(results)

        # Visualize comparison
        fig, axes = plt.subplots(2, 2, figsize=(15, 10))
        fig.suptitle('Filter Impact Analysis', fontsize=16, fontweight='bold')

        # Movie count comparison
        axes[0, 0].bar(comparison_df['Filter Name'], comparison_df['Movie Count'],
                       color=self.colors[:len(comparison_df)])
        axes[0, 0].set_title('Movie Count by Filter')
        axes[0, 0].set_ylabel('Number of Movies')
        plt.setp(axes[0, 0].get_xticklabels(), rotation=45)

        # Average rating comparison
        axes[0, 1].bar(comparison_df['Filter Name'], comparison_df['Avg Rating'],
                       color=self.colors[:len(comparison_df)])
        axes[0, 1].set_title('Average Rating by Filter')
        axes[0, 1].set_ylabel('Average Rating')
        plt.setp(axes[0, 1].get_xticklabels(), rotation=45)

        # Average revenue comparison
        axes[1, 0].bar(comparison_df['Filter Name'], comparison_df['Avg Revenue'],
                       color=self.colors[:len(comparison_df)])
        axes[1, 0].set_title('Average Revenue by Filter')
        axes[1, 0].set_ylabel('Average Revenue (Millions)')
        plt.setp(axes[1, 0].get_xticklabels(), rotation=45)

        # Summary table
        axes[1, 1].axis('off')
        table_data = comparison_df.round(2)
        table = axes[1, 1].table(cellText=table_data.values,
                                colLabels=table_data.columns,
                                cellLoc='center',
                                loc='center')
        table.auto_set_font_size(False)
        table.set_fontsize(8)
        table.scale(1.2, 1.5)
        axes[1, 1].set_title('Filter Comparison Summary')

        plt.tight_layout()
        return fig, comparison_df

# Example usage and demonstration
def main():
    # Initialize visualizer
    viz = MovieDataVisualizer()

    print("Movie Data Visualization System")
    print("="*50)

    # 1. Basic dashboard
    print("1. Creating basic matplotlib dashboard...")
    fig1 = viz.create_matplotlib_dashboard()
    plt.show()

    # 2. Advanced seaborn plots
    print("2. Creating advanced seaborn visualizations...")
    fig2 = viz.create_seaborn_advanced_plots()
    plt.show()

    # 3. Interactive Plotly dashboard
    print("3. Creating interactive Plotly dashboard...")
    plotly_fig = viz.create_plotly_interactive_dashboard()
    plotly_fig.show()

    # 4. Custom filtering examples
    print("4. Demonstrating custom filtering...")

    # High-quality movies filter
    high_quality_fig, high_quality_df = viz.create_custom_filter_visualization(
        min_rating=8.0,
        min_revenue=50,
        genres=['Action', 'Drama', 'Thriller']
    )
    plt.show()

    # Recent blockbusters filter
    blockbuster_fig, blockbuster_df = viz.create_custom_filter_visualization(
        min_year=2015,
        min_revenue=100,
        min_votes=50000
    )
    plt.show()

    # 5. Filter impact analysis
    print("5. Analyzing filter impact...")
    filter_sets = {
        'All Movies': {},
        'High Quality': {'min_rating': 8.0, 'min_revenue': 50},
        'Recent': {'min_year': 2020},
        'Blockbusters': {'min_revenue': 100, 'min_votes': 50000},
        'Action Movies': {'genres': ['Action']},
        'Premium': {'min_rating': 8.5, 'min_revenue': 200, 'min_votes': 75000}
    }

    impact_fig, comparison_df = viz.analyze_filter_impact(filter_sets)
    plt.show()

    print("\nFilter Impact Summary:")
    print(comparison_df)

    # 6. Export capabilities
    print("6. Saving visualizations...")

    # Save high-quality movies data
    high_quality_df.to_csv('high_quality_movies.csv', index=False)
    print("Saved high_quality_movies.csv")

    # Save plots
    high_quality_fig.savefig('high_quality_movies_analysis.png', dpi=300, bbox_inches='tight')
    print("Saved high_quality_movies_analysis.png")

    # Save interactive plot as HTML
    plotly_fig.write_html("interactive_movie_dashboard.html")
    print("Saved interactive_movie_dashboard.html")

if __name__ == "__main__":
    # For using with your own data:
    # df = pd.read_csv('your_movie_data.csv')
    # viz = MovieDataVisualizer(df)

    main()

# Additional utility functions for specific use cases
def quick_genre_analysis(df, genre):
    """Quick analysis for a specific genre"""
    genre_df = df[df['Genre'] == genre]

    print(f"\n{genre} Movies Analysis:")
    print(f"Total movies: {len(genre_df)}")
    print(f"Average rating: {genre_df['Rating'].mean():.2f}")
    print(f"Average revenue: ${genre_df['Revenue (Millions)'].mean():.1f}M")
    print(f"Year range: {genre_df['Year'].min()} - {genre_df['Year'].max()}")

    # Quick visualization
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))

    # Rating distribution
    axes[0].hist(genre_df['Rating'], bins=10, alpha=0.7, color='skyblue')
    axes[0].set_title(f'{genre} - Rating Distribution')
    axes[0].set_xlabel('Rating')

    # Revenue over time
    axes[1].scatter(genre_df['Year'], genre_df['Revenue (Millions)'], alpha=0.6)
    axes[1].set_title(f'{genre} - Revenue Over Time')
    axes[1].set_xlabel('Year')
    axes[1].set_ylabel('Revenue (Millions)')

    # Runtime vs Rating
    axes[2].scatter(genre_df['Runtime (Minutes)'], genre_df['Rating'], alpha=0.6, color='green')
    axes[2].set_title(f'{genre} - Runtime vs Rating')
    axes[2].set_xlabel('Runtime (Minutes)')
    axes[2].set_ylabel('Rating')

    plt.tight_layout()
    plt.show()

def compare_decades(df):
    """Compare movie statistics across decades"""
    df['Decade'] = (df['Year'] // 10) * 10
    decade_stats = df.groupby('Decade').agg({
        'Rating': 'mean',
        'Revenue (Millions)': 'mean',
        'Runtime (Minutes)': 'mean',
        'Title': 'count'
    }).round(2)
    decade_stats.columns = ['Avg Rating', 'Avg Revenue', 'Avg Runtime', 'Movie Count']

    print("\nDecade Comparison:")
    print(decade_stats)

    # Visualize decade trends
    fig, axes = plt.subplots(2, 2, figsize=(14, 10))

    decades = decade_stats.index

    axes[0, 0].plot(decades, decade_stats['Avg Rating'], marker='o', linewidth=2)
    axes[0, 0].set_title('Average Rating by Decade')
    axes[0, 0].set_ylabel('Rating')

    axes[0, 1].plot(decades, decade_stats['Avg Revenue'], marker='s', linewidth=2, color='green')
    axes[0, 1].set_title('Average Revenue by Decade')
    axes[0, 1].set_ylabel('Revenue (Millions)')

    axes[1, 0].plot(decades, decade_stats['Avg Runtime'], marker='^', linewidth=2, color='red')
    axes[1, 0].set_title('Average Runtime by Decade')
    axes[1, 0].set_ylabel('Runtime (Minutes)')

    axes[1, 1].bar(decades, decade_stats['Movie Count'], color='purple', alpha=0.7)
    axes[1, 1].set_title('Movie Count by Decade')
    axes[1, 1].set_ylabel('Number of Movies')

    plt.tight_layout()
    plt.show()

    return decade_stats
```
