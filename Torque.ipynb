```python
df.groupby('Product ID')['Air temperature [K]'].sum().value_counts(ascending=False).head(10)
```

```python
best_sales = df.groupby('Product ID')['Torque [Nm]'].sum().sort_values(ascending=False).head(10)

best_sales
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot showing the 10 products with the highest torque
plt.figure(figsize=(12,6))
sns.barplot(
    x=best_sales.values,
    y=best_sales.index,
    palette='viridis'
)

plt.title('Top 10 Products with the Highest Total Torque', fontsize=14)
plt.xlabel('Toplam Torque [Nm]', fontsize=12)
plt.ylabel('Product ID', fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
