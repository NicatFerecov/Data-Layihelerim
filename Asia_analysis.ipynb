```python
import pandas as pd
```

```python
df = pd.read_csv('/content/world_population.csv')
```

```python
df.info()
df.head(30)
```

```python
new_data = df[df['Continent'] == "Asia"]
new_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(new_data[['2022 Population', '2015 Population', '2000 Population',
                       '1980 Population', 'Area (km²)', 'Growth Rate']])
plt.suptitle('Population and Growth Relationship in ASIA Countries', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.heatmap(new_data[['2022 Population', '2015 Population', '2000 Population',
                      '1980 Population', 'Area (km²)', 'Growth Rate']].corr(),
            annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('ASIA Countries Variable Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.barplot(data=new_data.sort_values(by='2022 Population', ascending=False),
            x='2022 Population', y='Country', palette='viridis')
plt.title('ASIA Countries 2022 Population Ranking')
plt.xlabel('Population')
plt.ylabel('Country')
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(new_data['World Population Percentage'], labels=new_data['Country'],
        autopct='%1.1f%%', startangle=140)
plt.title('Share of ASIA Countries in World Population')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=new_data, x='Area (km²)', y='Density (per km²)', size='2022 Population',
                hue='Country', sizes=(20, 400))
plt.title('Area and Population Density Relationship in ASIA Countries')
plt.xlabel('Surface area (km²)')
plt.ylabel('Population Density (person/km²)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

# Conclusion
This notebook analyses the population size, growth rate, area and contribution of ASIA countries to the world population in the form of heat map, bar chart, pie chart, pairplot and scatter plot.
